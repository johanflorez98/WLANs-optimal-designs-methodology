```python
# Los nombres de los casos (puedes personalizarlos según tus necesidades)
casos = ["1 APs", "2 APs", "3 APs", "4 APs", "5 APs"]

# Genera cinco vectores RMSE de ejemplo, cada uno con 40 valores
np.random.seed(0)  # Esto es para reproducibilidad
rmse1 = mean_val_rmse_1*20
rmse2 = mean_val_rmse_2*20
rmse3 = mean_val_rmse_3*20
rmse4 = mean_val_rmse_4*20
rmse5 = mean_val_rmse_5*20

# Organizar los datos en un array bidimensional (40 filas x 5 columnas)
rmse_data = np.array([rmse1, rmse2, rmse3, rmse4, rmse5]).T

# Crear un array con los índices de los casos (1, 2, 3, ..., 40)
indices = np.arange(1, 41)

# Crear la gráfica de líneas
plt.figure(figsize=(8, 5))  # Tamaño de la figura
for i, caso in enumerate(casos):
    plt.plot(indices, rmse_data[:, i], marker='o', label=caso)

# Personalizar la gráfica
plt.xlabel("Epochs")
plt.ylabel("RMSE [m]")
plt.title("RMSE to APs cases validation")
plt.legend()
plt.grid(True)

# Establecer la fuente y el tamaño
plt.rc('font', family='Times New Roman', size=15)

# Mostrar la gráfica
plt.show()
```
